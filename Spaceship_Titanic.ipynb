```python
import zipfile
import os
import pandas as pd

# Ruta al archivo ZIP
archivo_zip = '/content/drive/MyDrive/Kaggle competitions/Spaceship Titanic/spaceship-titanic.zip'

# Ruta al directorio de extracción
directorio_extraccion = '/content/drive/MyDrive/Kaggle competitions/Spaceship Titanic'

# Descomprimir el archivo ZIP
with zipfile.ZipFile(archivo_zip, 'r') as zip_ref:
    zip_ref.extractall(directorio_extraccion)

# Lista de archivos en el directorio extraído
archivos_extraidos = os.listdir(directorio_extraccion)

# Diccionario para almacenar los DataFrames
dataframes = {}

# Crear un DataFrame para cada archivo CSV
for archivo in archivos_extraidos:
    if archivo.endswith('.csv'):
        ruta_archivo = os.path.join(directorio_extraccion, archivo)
        # Leer el archivo CSV y almacenar en el diccionario
        df = pd.read_csv(ruta_archivo)
        dataframes[archivo] = df
        # print(f"\nContenido de {archivo}:\n{'='*30}")
        # print(df.head())

# Ahora puedes acceder a los DataFrames por sus nombres de archivo
# Por ejemplo, para acceder al DataFrame 'sample_submission.csv':
sample_submission = dataframes.get('sample_submission.csv')
test_df           = dataframes.get('test.csv')
train_df          = dataframes.get('train.csv')

```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, HistGradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score

# Copiar train_df para evitar cambios en el DataFrame original
df = train_df.copy()

# Definir las características y la variable objetivo
X = df.drop(['PassengerId', 'Transported'], axis=1)
y = df['Transported']

# Dividir los datos en conjuntos de entrenamiento y validación
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.21, random_state=42)

# Definir transformadores para características numéricas y categóricas
numeric_features = X.select_dtypes(include=['number']).columns
categorical_features = X.select_dtypes(include=['object']).columns

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combinar transformadores
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Construir modelos de clasificación
random_forest_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42))
])

gradient_boosted_trees_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', GradientBoostingClassifier(random_state=42))
])

cart_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', DecisionTreeClassifier(random_state=42))
])

distributed_gradient_boosted_trees_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', HistGradientBoostingClassifier(random_state=42, max_iter=100))
])

# Entrenar modelos
random_forest_model.fit(X_train, y_train)
gradient_boosted_trees_model.fit(X_train, y_train)
cart_model.fit(X_train, y_train)
#distributed_gradient_boosted_trees_model.fit(X_train, y_train)

# Realizar predicciones en el conjunto de validación
predictions_rf = random_forest_model.predict(X_val)
predictions_gb = gradient_boosted_trees_model.predict(X_val)
predictions_cart = cart_model.predict(X_val)
#predictions_distributed_gb = distributed_gradient_boosted_trees_model.predict(X_val)

# Evaluar el rendimiento de los modelos
accuracy_rf = accuracy_score(y_val, predictions_rf)
accuracy_gb = accuracy_score(y_val, predictions_gb)
accuracy_cart = accuracy_score(y_val, predictions_cart)
#accuracy_distributed_gb = accuracy_score(y_val, predictions_distributed_gb)

print(f'Accuracy Random Forest: {accuracy_rf}')
print(f'Accuracy Gradient Boosted Trees: {accuracy_gb}')
print(f'Accuracy CART (Decision Tree): {accuracy_cart}')
#print(f'Accuracy Distributed Gradient Boosted Trees: {accuracy_distributed_gb}')
```
