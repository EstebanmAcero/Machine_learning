## Vamos a entrenar el modelo de manera que le demos paramentros

```python


##iris_train, iris_test = train_test_split(iris_df, train_size = 0.6, test_size= 0.4, shuffle= True) # suffle mezclar
iris_train, iris_test = train_test_split(iris_df, train_size = 0.29,  shuffle= True) # suffle mezclar


print(iris_train.shape)

# tenemos 4 features y 1 samples

iris_train.tail()

# cuando llamo a una porcion del dataframe x son solo features
# cuando llamo a una porcion del dataframe y son el target

iris_train_x = iris_train.drop("Target", axis = "columns")
iris_train_y = iris_train["Target"]

iris_test_x = iris_test.drop("Target", axis = "columns")
iris_test_y = iris_test["Target"]
```

## Modelo de Valle gausiano
Es un algoritmo de clasificacion

```python
from sklearn.naive_bayes import GaussianNB
model  = GaussianNB()

# Entrenarlo
model.fit(iris_train_x, iris_train_y) # fuatures y luego target


```

```python
# Vamos a comprobar una inferencia
# Como resulta la predicción después de entrenar el modelo

inference = model.predict(iris_test_x)
# vamos a comprobar que tan bien lo hace con el test
print(inference) # respuesta que da la máquina vs las respuestas verdaderas
# hacer las metricas de rendimiento


```

```python
# hacer las metricas de rendimiento
from sklearn.metrics import accuracy_score

model_acurracy = accuracy_score(iris_test_y, inference)

# la exactitud del modelo es

print("La exactitud del modelo es de: ", str(model_acurracy*100))
```

### Implementar un algoritmo no supervisado como clusterizacion para poder discriminar las tres categorias

Modelo
Gaussian mixed model

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.mixture import GaussianMixture

# URL del conjunto de datos Iris en formato CSV
iris_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

# Nombres de las columnas para el conjunto de datos Iris
column_names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "Target"]

# Cargar el conjunto de datos desde la URL en un DataFrame de pandas
iris_df = pd.read_csv(iris_url, names=column_names)


```

```python
import numpy as np
from sklearn.mixture import GaussianMixture
X = np.array([[1, 2], [1, 4], [1, 0], [10, 2], [10, 4], [10, 0]])
print(X)
gm = GaussianMixture(n_components=2, random_state=0).fit(X)
print(gm.means_)

gm.predict([[0, 0],
            [12, 3]])
```

```python
# Importa las bibliotecas necesarias
# Selecciona las características para el clustering (en este caso, todas las columnas excepto 'Target')
X = iris_df.drop('Target', axis=1)

# Define el número de clusters que deseas encontrar
n_clusters = 3

# Crea una instancia del modelo de mezcla gaussiana
gmm = GaussianMixture(n_components=n_clusters)

# Ajusta el modelo a tus datos
gmm.fit(X)

# Predice los clusters para cada punto de datos
y_cluster = gmm.predict(X)

# Agrega las etiquetas de cluster al DataFrame original
iris_df['Cluster'] = y_cluster

# Visualización de los resultados
sns.set_style("whitegrid")
sns.pairplot(iris_df, hue="Cluster", palette="Dark2")
sns.pairplot(iris_df, hue="Target", palette="Dark2")

plt.show()
print(iris_df)


```

## Analisis Grafico del codo

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Cargar el conjunto de datos de iris
iris = iris_df['sepal_length']
x = iris.values.reshape(-1, 1)  # Reformatear a un arreglo bidimensional

# Calcular la variación intra-cluster para diferentes valores de k
inertia = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(x)
    inertia.append(kmeans.inertia_)

# Graficar el análisis de codo
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Número de Clusters')
plt.ylabel('Variación Intra-Cluster (Inertia)')
plt.title('Análisis de Codo para K-Means')
plt.show()

```
